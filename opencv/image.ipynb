```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Create a hard-coded 2x2 matrix for a 4-pixel image
# The dimensions are (height=2, width=2, channels=3)
# The order of colors is BGR (Blue, Green, Red)

"""
This is the default color space for images read and stored by OpenCV. 
It's very similar to the RGB model used in digital displays but with the order of the first two channels swapped.
"""
four_pixel_matrix = np.array([
    [[100, 0, 0], [0, 255, 0], [0, 150, 10]], 
    [[0, 0, 255], [255, 255, 0], [0,0,0]]  
], dtype=np.uint8)

# Convert the color space from BGR to RGB for Matplotlib
# image_rgb = cv2.cvtColor(four_pixel_matrix, cv2.COLOR_BGR2GRAY)

# Display the image using Matplotlib
# The interpolation='nearest' argument is used to prevent blurring,
# which is essential for viewing individual pixels.
plt.imshow(four_pixel_matrix)
plt.axis('off')  # Hide the axes
plt.show()
```

```python
rgb_image = four_pixel_matrix[:, :, ::-1]

# 3. Display the RGB image with Matplotlib
plt.imshow(rgb_image)
plt.title("Manually Converted BGR to RGB")
plt.axis('off')  # Hide axes
plt.show()
```
